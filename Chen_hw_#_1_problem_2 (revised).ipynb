```python
import scipy as sp
import numpy as np
from scipy import special
import matplotlib.pyplot as plt
def approxairy(x):
    return np.exp(-2.0*(x**(1.5))/3.0)/(2.0*np.sqrt(sp.pi)*(x**0.25))
x=np.linspace(1,50,101)
plt.figure(figsize=(4,3),dpi=98)
label_a = 'The approximate airy function'
label_b = 'The exact airy function'
label_c = 'The error'
plt.plot(x,approxairy(x),"b",label=label_a,linewidth=1)
plt.plot(x,sp.special.airy(x)[0],"r",label=label_b,linewidth=1)
#plt.plot(x,(approxairy(x)-sp.special.airy(x)[0])/sp.special.airy(x)[0],"g",linewidth=1)
#sp.special.airy(x)[0]
#ax.set_xscale('log')
#ax.set_yscale('log')
plt.grid(True)
#plt.axis([0,1.6,0,0.071])
#plt.xscale('log')
#plt.yscale('log')
plt.legend()
plt.tick_params(labelsize=10)
#plt.setp(plt.gca().get_legend().get_texts(), fontsize='9')
plt.xlabel('x',fontsize=10)
#plt.ylabel('Relative Error ',fontsize=10)
#print(np.log(x))
plt.show()
plt.figure(figsize=(4,3),dpi=98)
#label_a = 'The approximate airy function'
#label_b = 'The exact airy function'
label_c = 'The error'
#plt.plot(x,approxairy(x),"b",label=label_a,linewidth=1)
#plt.plot(x,sp.special.airy(x)[0],"r",label=label_b,linewidth=1)
plt.plot(x,(approxairy(x)-sp.special.airy(x)[0])/sp.special.airy(x)[0]
         ,"g",linewidth=1)
#sp.special.airy(x)[0]
#ax.set_xscale('log')
#ax.set_yscale('log')
plt.grid(True)
#plt.axis([0,1.6,0,0.071])
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.tick_params(labelsize=10)
#plt.setp(plt.gca().get_legend().get_texts(), fontsize='9')
plt.xlabel('x',fontsize=10)
plt.ylabel('Relative Error (%)',fontsize=10)
#print(np.log(x))
plt.show()
for y in range(101):
    print('Relative Error ',(approxairy(x)[y]-sp.special.airy(x)[0][y])/sp.special.airy(x)[0][y],"%, When x=",x[y])
```
